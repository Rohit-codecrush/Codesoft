```python
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
data=r"C:\Users\admin\Desktop\IRIS.csv"
df=pd.read_csv(data)
```

```python
df
```

```python
df['species'].replace({'Iris-setosa':'1','Iris-virginica':'2'},inplace=True)
df
```

```python
sns.heatmap(df.isnull())
plt.show()            
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(df[['sepal_length','petal_length','petal_width']],df['species'],test_size=0.3)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
model.predict(x_test)
```

```python
model.predict([[5.1,1.5,0.2]])
```
